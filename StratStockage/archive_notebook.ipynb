```python
# Definition de la classe Techno
class Techno:
    def __init__(self, name, stored, prod, etain, etaout, Q, S, Vol):
        self.name = name     # nom de la techno
        self.stored = stored # ce qui est stock
        self.prod = prod     # ce qui est produit
        self.etain = etain     # coefficient de rendement  la charge
        self.etaout = etaout   # coefficient de rendement  la dcharge
        self.Q=Q             # capacit installe (flux sortant)
        self.S=S             # capacit de charge d'une techno de stockage (flux entrant)
        self.vol=Vol         # Capacit maximale de stockage de la techno (Volume max)

#phs=Techno(np.zeros(8560),np.zeros(8560),0.9,0.9,9.3,9.3,180)
```

```python
# Definition des fonctions de charge et dcharge d'une technologie
def load(tec,k,astocker):
    if astocker==0:
        out=0
    else:
        temp=min(astocker*tec.etain,tec.vol-tec.stored[k-1],tec.S*tec.etain)
        tec.stored[k:]=tec.stored[k-1]+temp
        out=astocker-temp/tec.etain
    return out

def unload(tec,k,aproduire):
    if aproduire==0:
        out=0
    else:
        temp=min(aproduire/tec.etaout,tec.stored[k],tec.Q/tec.etaout)
        if tec.name=='Lake':
            tec.stored[k:int(endmonthlake[k])]=tec.stored[k]-temp
        else:
            tec.stored[k:]=tec.stored[k]-temp
        tec.prod[k]=temp*tec.etaout
        out=aproduire-tec.prod[k]
    return out


#!!!!!!!!!!!!!!!!!
# Centrales thermiques
def thermProd(tec, k, aproduire):
    temp=min(aproduire/tec.etaout,tec.Q/tec.etaout)
    tec.prod[k]=temp*tec.etaout
    out=aproduire-tec.prod[k]
    
    return out


#!!!!!!!!!!!!!!!!!

# Sur ATH : pas de pnurie avec 56 centrales min.
# Sur ATL : 28 centrales min. (50%)
# Diviser en 4 ou 8 groupes (plutt 8 pour les besoins humains)
# 1/8 = 0.125, 7/8 = 0.875
# 2180, 2920, 3650, 4400, 5130, 5900, 6732, 7580
# Tiers de 8760 : 2920(4*730), 5840, 8460
# DANS le dernier tiers : 50% croissance linaire min, 25% baisse de 20% prod min/max, 25% arrt

# La production nuclaire est divise en 8 groupes, chacun a son cycle d'arrt. 
# Dans cette fonction, on regarde dans quel partie du cycle on est pour chaque groupe, 
# pour calibrer la production min et max.
def cycle(k):
    N = 8
    n = 1/N
    
    # Intervalles des 3 parties importantes du cycle, pour chaque groupe
    A_ranges = [((2180+6570)%8760,(2180+8030)%8760), ((2920+6570)%8760,(2920+8030)%8760), 
               ((3650+6570)%8760,(3650+8030)%8760), ((4400+6570)%8760,(4400+8030)%8760),
               ((5130+6570)%8760,(5130+8030)%8760), ((5900+6570)%8760,(5900+8030)%8760),
               ((6732+6570)%8760,(6732+8030)%8760), ((7580+6570)%8760,(7580+8030)%8760)]
    
    B_ranges = [((2180+8030)%8760,2180), ((2920+8030)%8760,2920), ((3650+8030)%8760,3650),
               ((4400+8030)%8760,4400), ((5130+8030)%8760,5130), ((5900+8030)%8760,5900),
               ((6732+8030)%8760,6732), ((7580+8030)%8760,7580)]
    
    C_ranges = [(2180, (2180+730)%8760), (2920, (2920+730)%8760), (3650, (3650+730)%8760),
               (4400, (4400+730)%8760), (5130, (5130+730)%8760), (5900, (5900+730)%8760),
               (6732, (6732+730)%8760), (7580, (7580+730)%8760)]

    inA = [lower <= k < upper for (lower, upper) in A_ranges]
    inB = [lower <= k < upper for (lower, upper) in B_ranges]
    inC = [lower <= k < upper for (lower, upper) in C_ranges]
    
    sMin = 0
    sMax = 0
    
    # Pour chaque groupe, on regarde sa zone et on ajuste son min et son max
    for i in range(N):
        if inA[i]:
            start = A_ranges[i][0]
            sMin += n * (0.2 + 0.00054795*(k-start))
            sMax += n * 1
        elif inB[i]:
            start = B_ranges[i][0]
            sMin += n * (1 - 0.00027397260274*(k-start))
            sMax += n * (1 - 0.00027397260274*(k-start))
        elif inC[i]:
            sMin += n * 0
            sMax += n * 0
        else:
            sMin += n * 0.2
            sMax += n * 1
    
    return (sMin, sMax)


# Si la demande est trop faible ou nulle, on produit quand mme  hauteur de 20%
def nucProd(tec, k, aproduire):
    MinMax = cycle(k)
    Pmin = MinMax[0]
    Pmax = MinMax[1]
    
    if aproduire > tec.Q/tec.etaout * Pmin:
        temp=min(aproduire/tec.etaout,tec.Q*Pmax/tec.etaout)
        tec.prod[k]=temp*tec.etaout
    else:
        tec.prod[k] = tec.Q/tec.etaout * Pmin
    
    out = aproduire - tec.prod[k]
    
    return out
```

```python
def StratStockage(prodres,n,Phs,Battery,Methanation,Lake,Thermal,Nuclear):
    Surplus=np.zeros(n)
    ##Ajout paramtre Penurie
    Manque = np.zeros(n)
    #Definition d'un ordre sur les differentes technologies de stockage et destockage
    Tecstock= {"Phs" : Phs,"Battery" : Battery, "Methanation": Methanation}
    Tecstock2= {"Methanation": Methanation,"Phs" : Phs,"Battery" : Battery}
        
    Tecdestock= {"Battery" : Battery,"Phs" : Phs,"Methanation": Methanation,"Lake" : Lake}
    
    for k in np.arange(1,n):
        if prodres[k]>0:
            
            #!!!!!!!!!!!!!!!!!!!!!
            # La production min de nuclaire s'ajoute  la qt d'nergie  stocker
            nucMin = nucProd(Nuclear, k, 0)
            Astocker = prodres[k] + abs(nucMin)
            
            for tec in Tecstock:
                Astocker=load(Tecstock[tec],k,Astocker)
            Surplus[k]=Astocker
        else:
            Aproduire=-prodres[k]
            
            #!!!!!!!!!!!!!!!!!!!
            Aproduire = nucProd(Nuclear, k, Aproduire)
            
            for tec in Tecdestock:
                Aproduire=unload(Tecdestock[tec],k,Aproduire)
                
            #!!!!!!!!!!!!!!!!!!!
            # Si le nuclaire n'a pas suffi, on fait tourner les centrales thermiques
            if Aproduire > 0:
                Aproduire = thermProd(Thermal, k, Aproduire)
                
            ##liste penurie --> pour savoir si il y a pnurie dans la production d'lectricit 
            Manque[k]=Aproduire
                
    return Surplus,Manque,Phs,Battery,Methanation,Lake,Thermal,Nuclear

def StratStockagev2(prodres,n,Phs,Battery,Methanation,Lake):
    Ratio=0
    Surplus=np.zeros(n)
    #Definition d'un ordre sur les differentes technologies de stockage et destockage
    Tecstock= {"Phs" : Phs,"Battery" : Battery, "Methanation": Methanation}
    Tecstock2= {"Methanation": Methanation,"Phs" : Phs,"Battery" : Battery}      
    #Tecdestock= {"Phs" : Phs,"Battery" : Battery, "Methanation": Methanation}
    Tecdestock= {"Battery" : Battery,"Phs" : Phs,"Lake" : Lake, "Methanation" : Methanation}
    
    for k in np.arange(1,n):
        if prodres[k]>0:
            Astocker=prodres[k]
            for tec in Tecstock:
                Astocker=load(Tecstock[tec],k,Astocker)
                #print(Astocker,Tecstock[tec].stored[k])
            Surplus[k]=Astocker
        else:
            Aproduire=-prodres[k]
            ####### A Completer 
            if (Aproduire<Phs.Q+Battery.Q) & (Phs.stored[k]+Battery.stored[k]>Ratio*(Phs.vol+Battery.vol)):
                Aproduire2=min(Aproduire+Methanation.S,Phs.Q+Battery.Q)
                Aproduire3=load(Methanation,k,Aproduire2-Aproduire)
                Aproduire=unload(Phs,k,Aproduire2)
                Aproduire=unload(Battery,k,Aproduire)
            else:
                for tec in Tecdestock:
                    Aproduire=unload(Tecdestock[tec],k,Aproduire)
    return Surplus,Phs,Battery,Methanation,Lake


def StratStockagev3(prodres,n,Phs,Battery,Methanation,Lake,Thermal,Nuclear, I0, I1, I2):
    Surplus=np.zeros(n)
    ##Ajout paramtre Penurie
    Manque = np.zeros(n)
    
    #Definition d'un ordre sur les differentes technologies de stockage et destockage
    Tecstock2= {"Methanation": Methanation,"Phs" : Phs,"Battery" : Battery} ##on stocke du gaz zone 1,2
    Tecstock3= {"Phs" : Phs,"Battery" : Battery, "Methanation": Methanation} ## zone 3
    Tecstock4 = {"Battery" : Battery, "Phs" : Phs,"Methanation": Methanation} ## zone 4
        
    Tecdestock1= {"Battery" : Battery,"Phs" : Phs,"Methanation": Methanation,"Lake" : Lake} #zone 1
    Tecdestock2 = {"Phs": Phs, "Battery": Battery, "Methanation" : Methanation, "Lake" : Lake} ## zone 2
    Tecdestock3 = {"Methanation" : Methanation, "Battery" : Battery, "Phs": Phs, "Lake" : Lake} ## zone 3,4
    
    for k in np.arange(1,n):
        stock_PB = Phs.stored[k-1] + Battery.stored[k-1]
        action = -1
        
        if 0 <= stock_PB < I0[k%24] :
            strat_stock = Tecstock4
            strat_destock = Tecdestock3
            qteInit = min(Methanation.Q, Phs.S+Battery.S)
            reste = unload(Methanation, k, qteInit)
            reste = load(Battery, k, qteInit-reste)
            load(Phs, k, reste)
            
        elif I0[k%24] <= stock_PB < I1[k%24] :
            strat_stock = Tecstock3
            strat_destock = Tecdestock3
            
        elif I1[k%24] <= stock_PB < I2[k%24] :
            strat_stock = Tecstock2
            strat_destock = Tecdestock2
            
        else :
            strat_stock = Tecstock2
            strat_destock = Tecdestock1
            qteInit = min(Phs.Q+Battery.Q, Methanation.S)
            reste = unload(Battery, k, qteInit)
            reste = unload(Phs, k, reste)
            load(Methanation, k, qteInit-reste)
            
            
        
        if prodres[k]>0 :
            #!!!!!!!!!!!!!!!!!!!!!
            # La production min de nuclaire s'ajoute  la qt d'nergie  stocker
            nucMin = nucProd(Nuclear, k, 0)
            Astocker = prodres[k] + abs(nucMin)
            
            for tec in strat_stock:
                Astocker=load(strat_stock[tec],k,Astocker)
            Surplus[k]=Astocker
        else:
            Aproduire=-prodres[k]

            #!!!!!!!!!!!!!!!!!!!
            Aproduire = nucProd(Nuclear, k, Aproduire)

            for tec in strat_destock:
                Aproduire=unload(strat_destock[tec],k,Aproduire)

            #!!!!!!!!!!!!!!!!!!!
            # Si le nuclaire n'a pas suffi, on fait tourner les centrales thermiques
            if Aproduire > 0:
                Aproduire = thermProd(Thermal, k, Aproduire)

            ##liste penurie --> pour savoir si il y a pnurie dans la production d'lectricit 
            Manque[k]=Aproduire
            
                
    return Surplus,Manque,Phs,Battery,Methanation,Lake,Thermal,Nuclear
```

```python
##Mthode Aymeric
def certitude1(y1,y2,y3,seuil):
    
    y = np.concatenate((y1,y2,y3), axis = 0)
    best_ratio = -1
    best_stock = 0
    
    for crit in range(270):
        nbr_Pen = len(y2[y2>crit])
        nbr = len(y[y>crit])
    
        if nbr !=0 :
            ratio = nbr_Pen/nbr
        if abs(ratio-seuil)<abs(best_ratio-seuil) :
            best_ratio = ratio
            best_stock = crit
    
    return best_stock

def certitude2(y1,y2,y3,seuil):
    y = np.concatenate((y1,y2,y3), axis = 0)
    best_ratio = -1
    best_stock = 0
    y1=y1[y1!=-1]
    y2=y2[y2!=-1]
    y3=y3[y3!=-1]
    
    for crit in range(270):
        nbr_Pen = len(y2[y2<crit])
        nbr = len(y[y<crit])
    
        if nbr !=0 :
            ratio = nbr_Pen/nbr
        if abs(ratio-seuil)<abs(best_ratio-seuil) :
            best_ratio = ratio
            best_stock = crit
    
    return best_stock

def certitudeglobal(y1,y2,y3):
    
        ##distribution cretage : min, max, moyenne et cart-type 
        emoy = np.mean(y1[y1!=-1]) ##moyenne de l'chantillon //
        eetype = np.std(y1[y1!=-1]) ##ecart-type de l'chantillon //
       
        ##distribution pnurie : min, max, moyenne, cart-type
        pmoy = np.mean(y2[y2!=-1])
        petype = np.std(y2[y2!=-1])
        
        certitude_interval = np.zeros(3)
        certitude_interval[0] = pmoy + (1.76)*(petype)/np.sqrt(len(y2[y2!=-1])) ##98% sur pnurie (valeur sup de l'IC)
        certitude_interval[1] = emoy - (2.33)*(eetype)/np.sqrt(len(y1[y1!=-1])) ##99% sur cretage (valeur inf de l'IC)
        certitude_interval[2] = (certitude_interval[0] + certitude_interval[1])/2 ##valeur moyenne entre 98% et 99% 
        
        return certitude_interval
    
```

```python
# Ecretage / Penurie / Ok
def seuil(a,b,c, crit, mode):
        
    y1 = np.copy(a)
    y2 = np.copy(b)
    y3 = np.copy(c)
    
    
    for i in range (len(y1)):
        y3[i] = -1 if (y1[i]==y3[i] or y2[i]==y3[i]) else y3[i]
        
    
    bestRatio = -1
    bestStock = -1
    
    for s in range(270):
        nbPen = 0
        nbSeuil = 0
    
        for i in range (len(y1)):
            if mode == "u":
                if y1[i] >= s or y3[i] >= s:
                    nbSeuil += 1
                elif y2[i] >= s:
                    nbSeuil += 1
                    nbPen += 1
            else:
                if 0 <= y1[i] <= s or 0 <= y3[i] <= s:
                    nbSeuil += 1
                elif 0 <= y2[i] <= s:
                    nbSeuil += 1
                    nbPen += 1
        
        if nbSeuil != 0:
            ratio = nbPen / nbSeuil
            if abs(ratio-crit) < abs(bestRatio-crit):
                bestRatio = ratio
                bestStock = s
            #print("stock : {}, ratio : {}, cart par rapport  {} : {}".format(s,ratio,crit,abs(crit-ratio)))
                
    
    return bestStock
```
