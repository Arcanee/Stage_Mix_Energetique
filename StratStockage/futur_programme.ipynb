# C. Classes et mÃ©thodes importantes 

```python
# Definition de la classe Techno
class Techno:
    def __init__(self, name, stored, prod, etain, etaout, Q, S, Vol):
        self.name = name     # nom de la techno
        self.stored = stored # ce qui est stockÃ©
        self.prod = prod     # ce qui est produit
        self.etain = etain     # coefficient de rendement Ã  la charge
        self.etaout = etaout   # coefficient de rendement Ã  la dÃ©charge
        self.Q=Q             # capacitÃ© installÃ©e (flux sortant)
        self.S=S             # capacitÃ© de charge d'une techno de stockage (flux entrant)
        self.vol=Vol         # CapacitÃ© maximale de stockage de la techno (Volume max)
```

```python
# Definition des fonctions de charge et dÃ©charge d'une technologie
def load(tec,k,astocker):
    if astocker==0:
        out=0
    else:
        temp=min(astocker*tec.etain,tec.vol-tec.stored[k-1],tec.S*tec.etain)
        tec.stored[k:]=tec.stored[k-1]+temp
        out=astocker-temp/tec.etain
    return out

def unload(tec,k,aproduire, prod=True):
    if aproduire==0:
        out=0
    else:
        temp=min(aproduire/tec.etaout,tec.stored[k],tec.Q/tec.etaout)
        if tec.name=='Lake':
            tec.stored[k:int(endmonthlake[k])]=tec.stored[k]-temp
        else:
            tec.stored[k:]=tec.stored[k]-temp
        if prod:
            tec.prod[k]=temp*tec.etaout
        out=aproduire-temp*tec.etaout
    return out
```

### Productions thermique et nuclÃ©aire 

```python
# Sur ATH : pas de pÃ©nurie avec 56 centrales min.
# Sur ATL : 28 centrales min. (50%)
# Diviser en 4 ou 8 groupes (plutÃ´t 8 pour les besoins humains)
# 1/8 = 0.125, 7/8 = 0.875
# 2180, 2920, 3650, 4400, 5130, 5900, 6732, 7580
# Tiers de 8760 : 2920(4*730), 5840, 8460
# DANS le dernier tiers : 50% croissance linÃ©aire min, 25% baisse de 20% prod min/max, 25% arrÃªt

# La production nuclÃ©aire est divisÃ©e en 8 groupes, chacun a son cycle d'arrÃªt. 
# Dans cette fonction, on regarde dans quel partie du cycle on est pour chaque groupe, 
# pour calibrer la production min et max.
def cycle(k):
    N = 8
    n = 1/N
    
    # Intervalles des 3 parties importantes du cycle, pour chaque groupe
    A_ranges = [((2180+6570)%8760,(2180+8030)%8760), ((2920+6570)%8760,(2920+8030)%8760), 
               ((3650+6570)%8760,(3650+8030)%8760), ((4400+6570)%8760,(4400+8030)%8760),
               ((5130+6570)%8760,(5130+8030)%8760), ((5900+6570)%8760,(5900+8030)%8760),
               ((6732+6570)%8760,(6732+8030)%8760), ((7580+6570)%8760,(7580+8030)%8760)]
    
    B_ranges = [((2180+8030)%8760,2180), ((2920+8030)%8760,2920), ((3650+8030)%8760,3650),
               ((4400+8030)%8760,4400), ((5130+8030)%8760,5130), ((5900+8030)%8760,5900),
               ((6732+8030)%8760,6732), ((7580+8030)%8760,7580)]
    
    C_ranges = [(2180, (2180+730)%8760), (2920, (2920+730)%8760), (3650, (3650+730)%8760),
               (4400, (4400+730)%8760), (5130, (5130+730)%8760), (5900, (5900+730)%8760),
               (6732, (6732+730)%8760), (7580, (7580+730)%8760)]

    inA = [lower <= k < upper for (lower, upper) in A_ranges]
    inB = [lower <= k < upper for (lower, upper) in B_ranges]
    inC = [lower <= k < upper for (lower, upper) in C_ranges]
    
    sMin = 0
    sMax = 0
    
    # Pour chaque groupe, on regarde sa zone et on ajuste son min et son max
    for i in range(N):
        if inA[i]:
            start = A_ranges[i][0]
            sMin += n * (0.2 + 0.00054795*(k-start))
            sMax += n * 1
        elif inB[i]:
            start = B_ranges[i][0]
            sMin += n * (1 - 0.00027397260274*(k-start))
            sMax += n * (1 - 0.00027397260274*(k-start))
        elif inC[i]:
            sMin += n * 0
            sMax += n * 0
        else:
            sMin += n * 0.2
            sMax += n * 1
    
    return (sMin, sMax)


# Si la demande est trop faible ou nulle, on produit quand mÃªme Ã  hauteur de 20%
def nucProd(tec, k, aproduire):
    MinMax = cycle(k)
    Pmin = MinMax[0]
    Pmax = MinMax[1]
    
    if aproduire > tec.Q/tec.etaout * Pmin:
        temp=min(aproduire/tec.etaout,tec.Q*Pmax/tec.etaout)
        tec.prod[k]=temp*tec.etaout
    else:
        tec.prod[k] = tec.Q/tec.etaout * Pmin
    
    out = aproduire - tec.prod[k]
    
    return out



# Centrales thermiques
def thermProd(tec, k, aproduire):
    temp=min(aproduire/tec.etaout,tec.Q/tec.etaout)
    tec.prod[k]=temp*tec.etaout
    out=aproduire-tec.prod[k]
    
    return out
```

### TroisiÃ¨me mix : stratÃ©gie de stockage dÃ©pendant des stocks, dÃ©finition des seuils

```python
# MÃ©thode 1 : intervalles de confiance
def certitudeglobal(y1,y2,y3):
    
    ##distribution Ã©cretage : min, max, moyenne et Ã©cart-type 
    emoy = np.mean(y1[y1!=-1]) ##moyenne de l'Ã©chantillon //
    eetype = np.std(y1[y1!=-1]) ##ecart-type de l'Ã©chantillon //

    ##distribution pÃ©nurie : min, max, moyenne, Ã©cart-type
    pmoy = np.mean(y2[y2!=-1])
    petype = np.std(y2[y2!=-1])

    certitude_interval = np.zeros(3)
    certitude_interval[0] = pmoy + (1.76)*(petype)/np.sqrt(len(y2[y2!=-1])) ##98% sur pÃ©nurie (valeur sup de l'IC)
    certitude_interval[1] = emoy - (2.33)*(eetype)/np.sqrt(len(y1[y1!=-1])) ##99% sur Ã©cretage (valeur inf de l'IC)
    certitude_interval[2] = (certitude_interval[0] + certitude_interval[1])/2 ##valeur moyenne entre 98% et 99% 

    return certitude_interval

    

# MÃ©thode 2 : recherche itÃ©rative du meilleur seuil
# Cette fonction regarde la proportion de pÃ©nuries parmi les points au-dessus/sous du seuil, pour chaque seuil
# Et choisit le seuil Ã  la proportion la plus proche du critÃ¨re voulu

# Ecretage / Penurie / Ok
def seuil(a,b,c, crit, mode):
        
    y1 = np.copy(a)
    y2 = np.copy(b)
    y3 = np.copy(c)
    
    
    for i in range (len(y1)):
        y3[i] = -1 if (y1[i]==y3[i] or y2[i]==y3[i]) else y3[i]
        
    
    bestRatio = -1
    bestStock = -1
    
    for s in range(270):
        nbPen = 0
        nbSeuil = 0
    
        for i in range (len(y1)):
            if mode == "u":
                if y1[i] >= s or y3[i] >= s:
                    nbSeuil += 1
                elif y2[i] >= s:
                    nbSeuil += 1
                    nbPen += 1
            else:
                if 0 <= y1[i] <= s or 0 <= y3[i] <= s:
                    nbSeuil += 1
                elif 0 <= y2[i] <= s:
                    nbSeuil += 1
                    nbPen += 1
        
        if nbSeuil != 0:
            ratio = nbPen / nbSeuil
            if abs(ratio-crit) < abs(bestRatio-crit):
                bestRatio = ratio
                bestStock = s
                
    
    return bestStock
```

```python
def StratStockage(prodres,n,Phs,Battery,Methanation,Lake,Thermal,Nuclear):
    Surplus=np.zeros(n)
    ##Ajout paramÃ¨tre Penurie
    Manque = np.zeros(n)
    #Definition d'un ordre sur les differentes technologies de stockage et destockage
    Tecstock= {"Phs" : Phs,"Battery" : Battery, "Methanation": Methanation}
    Tecstock2= {"Methanation": Methanation,"Phs" : Phs,"Battery" : Battery}
        
    Tecdestock= {"Battery" : Battery,"Phs" : Phs,"Methanation": Methanation,"Lake" : Lake}
    
    for k in np.arange(1,n):
        if prodres[k]>0:
            
            # La production min de nuclÃ©aire s'ajoute Ã  la qtÃ© d'Ã©nergie Ã  stocker
            nucMin = nucProd(Nuclear, k, 0)
            Astocker = prodres[k] + abs(nucMin)
            
            for tec in Tecstock:
                Astocker=load(Tecstock[tec],k,Astocker)
            Surplus[k]=Astocker
        else:
            Aproduire=-prodres[k]
            
            Aproduire = nucProd(Nuclear, k, Aproduire)
            
            for tec in Tecdestock:
                Aproduire=unload(Tecdestock[tec],k,Aproduire)
                
            # Si le nuclÃ©aire n'a pas suffi, on fait tourner les centrales thermiques
            if Aproduire > 0:
                Aproduire = thermProd(Thermal, k, Aproduire)
                
            ##liste penurie --> pour savoir si il y a pÃ©nurie dans la production d'Ã©lectricitÃ© 
            Manque[k]=Aproduire
                
    return Surplus,Manque




def StratStockagev3(prodres,n,Phs,Battery,Methanation,Lake,Thermal,Nuclear, I0, I1, I2):
    Surplus=np.zeros(n)
    ##Ajout paramÃ¨tre Penurie
    Manque = np.zeros(n)
    
    #Definition d'un ordre sur les differentes technologies de stockage et destockage
    Tecstock2= {"Methanation": Methanation,"Phs" : Phs,"Battery" : Battery} ##on stocke du gaz zone 1,2
    Tecstock3= {"Phs" : Phs,"Battery" : Battery, "Methanation": Methanation} ## zone 3
    Tecstock4 = {"Battery" : Battery, "Phs" : Phs,"Methanation": Methanation} ## zone 4
        
    Tecdestock1= {"Battery" : Battery,"Phs" : Phs,"Methanation": Methanation,"Lake" : Lake} #zone 1
    Tecdestock2 = {"Phs": Phs, "Battery": Battery, "Methanation" : Methanation, "Lake" : Lake} ## zone 2
    Tecdestock3 = {"Methanation" : Methanation, "Battery" : Battery, "Phs": Phs, "Lake" : Lake} ## zone 3,4
    
    for k in np.arange(1,n):
        stock_PB = Phs.stored[k-1] + Battery.stored[k-1]
        
        # Suivant le niveau de stock, on change l'ordre de dÃ©/stockage et on fait du power2gaz ou
        # gaz2power si besoin
        
        if 0 <= stock_PB < I0[k%24] :
            strat_stock = Tecstock4
            strat_destock = Tecdestock3
            qteInit = min(Methanation.Q, Phs.S+Battery.S)
            reste = unload(Methanation, k, qteInit, False)
            reste = load(Battery, k, qteInit-reste)
            load(Phs, k, reste)
            
        elif I0[k%24] <= stock_PB < I1[k%24] :
            strat_stock = Tecstock3
            strat_destock = Tecdestock3
            
        elif I1[k%24] <= stock_PB < I2[k%24] :
            strat_stock = Tecstock2
            strat_destock = Tecdestock2
            
        else :
            strat_stock = Tecstock2
            strat_destock = Tecdestock1
            qteInit = min(Phs.Q+Battery.Q, Methanation.S)
            reste = unload(Battery, k, qteInit, False)
            reste = unload(Phs, k, reste, False)
            load(Methanation, k, qteInit-reste)
            
            
        
        if prodres[k]>0 :
            # La production min de nuclÃ©aire s'ajoute Ã  la qtÃ© d'Ã©nergie Ã  stocker
            nucMin = nucProd(Nuclear, k, 0)
            Astocker = prodres[k] + abs(nucMin)
            
            for tec in strat_stock:
                Astocker=load(strat_stock[tec],k,Astocker)
            Surplus[k]=Astocker
            
        else:
            Aproduire=-prodres[k]

            Aproduire = nucProd(Nuclear, k, Aproduire)

            for tec in strat_destock:
                Aproduire=unload(strat_destock[tec],k,Aproduire)

            # Si le nuclÃ©aire n'a pas suffi, on fait tourner les centrales thermiques
            if Aproduire > 0:
                Aproduire = thermProd(Thermal, k, Aproduire)
            
            ##liste penurie --> pour savoir si il y a pÃ©nurie dans la production d'Ã©lectricitÃ© 
            Manque[k]=Aproduire
            
                
    return Surplus,Manque
```

```python
## Etude d'Optimisation de stratÃ©gie de stockage et de dÃ©stockage du Mix Ã©nergetique 
class Mix3(param.Parameterized):
    Scenario_consom = param.ObjectSelector(default="ADEME",objects=Scenario.keys())
    onshore= param.Integer(75,bounds=(0,100))
    offshore = param.Integer(13,bounds=(0,100))
    
    thermique = param.Integer(0, bounds=(0,24))
    nucleaire = param.Integer(12, bounds=(0,56))
    
    pv= param.Integer(122,bounds=(0,150))
    nb= param.Integer(3,bounds=(1,52))
    prem=param.Integer(3,bounds=(1,52))
    H = param.Integer(0, bounds=(0,23)) ## ajout d'un paramÃ¨tre heure : permet d'Ã©tudier pour tous les jours les stocks Ã  l'heure H
    #heure_fin = param.Integer(23, bounds=(0,23))
    
    def opti_stock(self):
        # Choix du scenario de consommation par l'utilisateur
        Conso=Scenario[self.Scenario_consom]
        coefriv=13
        
        horaires=np.arange(1,8760)

        # Calcul de la production residuelle
        prodresiduelle=np.array(prod2006_offshore)*self.offshore\
                   +np.array(prod2006_onshore)*self.onshore\
                   +np.array(prod2006_pv)*self.pv+coefriv*rivprod-Conso
        
        # Puissance centrales territoire : 18.54 GWe rÃ©partis sur 24 centrales (EDF)
        # Rendement mÃ©ca (inutile ici) : ~35% gÃ©nÃ©ralement (Wiki)
        T = Techno('Centrale thermique', None, np.zeros(8760), None, 1, 0.7725*self.thermique, None, None)
        # Puissance : 1.08 GWe (EDF)
        # MÃªme rendement
        N = Techno('RÃ©acteur nuclÃ©aire', None, np.zeros(8760), None, 1, 1.08*self.nucleaire, None, None)
        
        # Definition des differentes technologies
        P=Techno('Phs', np.ones(8760)*90, np.zeros(8760), 0.95, 0.9, 9.3, 9.3, 180)
        B=Techno('Battery', np.ones(8760)*37.07, np.zeros(8760), 0.9, 0.95, 20.08, 20.08, 74.14)
        M=Techno('Methanation', np.ones(8760)*62500, np.zeros(8760), 0.59, 0.45, 32.93, 7.66, 125000)    
        L=Techno('Lake', storedlake, np.zeros(8760), 1, 1, 10, 10, 2000)   
        
        d=self.prem*7*24
        f=d+self.nb*7*24
            
        # rÃ©sultats de la strat initiale
        # Renvoie Surplus,PÃ©nurie,Phs,Battery,Methanation,Lake,Therm,Nuc
        s,p=StratStockage(prodresiduelle,8760,P,B,M,L,T,N)
        
        
        #############################
        ## NUAGES DE POINTS POUR OPTIMISER LE STOCKAGE
        
        stockage_PB = np.zeros(365) ##liste qui va servir Ã  enregister les stockages Phs + Battery Ã  l'heure H pour tous les jours
        
        stockmax=180+74.14 ##stockage maximum total = max total stockage Phs + max total stockage Battery
        x = np.arange(365) ##nombre de jours sur l'annÃ©e --> va de 1 Ã  365
        x0 = np.arange(8760)
        
        
        ##listes pour Ã©crÃªtage : x1 enregistre les jours oÃ¹ le lendemain il y a Ã©crÃªtage
        ##y1 enregistre la valeur du stock Phs + Battery oÃ¹ le lendemain il y a Ã©crÃªtage
        x1 = np.ones(365)*(-1)
        y1 = np.ones(365)*(-1)
        
        ##pareil que prÃ©cÃ¨demment mais pour lendemains avec pÃ©nurie
        x2 = np.ones(365)*(-1)
        y2 = np.ones(365)*(-1)
        
        ##pareil que prÃ©cÃ¨demment mais pour lendemains avec demande satisfaite et sans perte
        x3 = np.ones(365)*(-1)
        y3 = np.ones(365)*(-1)
        
        ##on enlevera les -1 des listes x1, x2, x3, y1, y2, y3 pour ne rÃ©cupÃ©rer que les points essentiels
            
        StockPB = P.stored + B.stored ##valeur des deux stocks 
        
      
        ###############################################################################
        ##Cetitude interval pour toutes les heures
        certitude_interval0 = np.zeros(24)
        certitude_interval1 = np.zeros(24)
        certitude_interval2 = np.zeros(24)
        
        seuils_top = np.zeros(24)
        seuils_mid = np.zeros(24)
        seuils_bot = np.zeros(24)
        
        for h in range(24):
        
            for jour in range(365): ##on regarde tous les jours de l'annÃ©e
            
                stockage_PB[jour]=StockPB[jour*24+h] #Au jour jour, valeur du stock Phs + Battery
            
                ##on regarde dans les 24h qui suivent si il y a Ã©crÃªtage, pÃ©nurie ou aucun des deux
                for heure in range(24): 
                    t = (jour * 24 + h + heure) % 8760
                    
                    if s[t] > 0 and StockPB[t] == stockmax : ##cas Ã©crÃªtage
                        x1[jour]=jour+1 ##on note le jour prÃ©cÃ¨dant jour avec Ã©crÃªtage
                        y1[jour]=stockage_PB[jour] ##on note le stock du jour prÃ©cÃ¨dant jour avec Ã©crÃªtage
                
                    elif p[t] > 0 : ##cas pÃ©nurie
                        x2[jour]=jour+1 ##mÃªmes explications mais pour pÃ©nurie
                        y2[jour]=stockage_PB[jour]
                    
                    else : ##cas ni Ã©crÃªtage, ni pÃ©nurie
                        x3[jour]=jour+1 ##mÃªmes explications mais avec ni Ã©crÃªtage, ni pÃ©nurie
                        y3[jour]=stockage_PB[jour]
                    
                    if x1[jour]==x2[jour]: ##si Ã©cretage et pÃ©nurie le mÃªme jour, on considÃ¨re que c'est une pÃ©nurie 
                        x1[jour]=-1
                        y1[jour]=-1
                
                
            certitude_interval0[h] = certitudeglobal(y1,y2,y3)[0]
            certitude_interval1[h] = certitudeglobal(y1,y2,y3)[1]
            certitude_interval2[h] = certitudeglobal(y1,y2,y3)[2]
            
            seuils_top[h] = seuil(y1,y2,y3, 0.02, "u")
            seuils_bot[h] = seuil(y1,y2,y3, 0.9, "d")
            seuils_mid[h] = (seuils_top[h] + seuils_bot[h]) / 2
            
            if h == 0:
                ##Courbe stockage Phs + Battery Ã  l'heure H tous les jours de l'annÃ©e
                C1 = hv.Curve((x,stockage_PB), 'jours', 'valeur du stock Phs + Battery (en GW)',
                              label = 'Stock Phs+Battery Ã  heure H tous les jours').options(width = 900, 
                                                                                            color = 'violet')

                C2 = hv.Curve((x0,s), 'heures', 'Valeur du surplus ou de la pÃ©nurie (en GW)',
                              label = 'Ecretage ou PÃ©nurie en fonction des heures').options(width = 900, 
                                                                                            color = 'orange')

                C3 = hv.Curve((x0,p), 'heures', 'Valeur du surplus ou de la pÃ©nurie (en GW)',
                              label = 'Ecretage ou PÃ©nurie en fonction des heures').options(width = 900, 
                                                                                            color = 'blue')


                ##Nuages de points indiquant si il y a Ã©crÃªtage, pÃ©nurie ou aucun des deux
                H1 = hv.Points((x1[x1!=-1],y1[y1!=-1]), kdims=['jours', 'valeur du stock Phs + Battery (en GW)'],
                               label = 'Ã©cretage').options(width = 900, color = 'orange', size = 14)

                H2 = hv.Points((x2[x2!=-1],y2[y2!=-1]), kdims=['jours', 'valeur du stock Phs + Battery (en GW)'],
                               label = 'pÃ©nurie').options(width = 900, color = 'blue', size = 10)

                H3 = hv.Points((x3[x3!=-1],y3[y3!=-1]), kdims=['jours', 'valeur du stock Phs + Battery (en GW)'],
                               label = 'ok').options(width = 900, color = 'green', size = 6, marker = 'square')


                ##Intervalles de confiance --> distributions pÃ©nurie et Ã©crÃªtage supposÃ©es normale ?
                certitude_interval = certitudeglobal(y1,y2,y3)


                ##Affichage de la zone de confiance sur le graphe : Nuages de points                               
                H4 = hv.Curve((x,certitude_interval[0]*np.ones(365)),
                              label = 'seuil de certitude inf').options(width = 900, color='red')#borne inf

                H5 = hv.Curve((x, certitude_interval[1]*np.ones(365)),
                              label = 'seuil de certitude sup').options(width=900, color = 'black') #borne sup

                H6 = hv.Curve((x, certitude_interval[2]*np.ones(365)),
                              label = 'valeur mÃ©diane').options(width = 900, color = 'cyan') #valeur mÃ©diane

                H7 = hv.Curve((x0, M.stored), "heures", "stock de gaz", label="stock gaz naif").options(width=900)
            
            
        
        xh = np.arange(24)
    
        G1 = hv.Curve((xh,certitude_interval0), 'Heures', 'valeur stock seuil',
                      label = 'seuil de certitude inf').options(width = 900, color='red')#borne inf
        
        G2 = hv.Curve((xh, certitude_interval1), 'Heures', 'valeur stock seuil',
                      label = 'seuil de certitude sup').options(width=900, color = 'black') #borne sup
        
        G3 = hv.Curve((xh, certitude_interval2), 'Heures', 'valeur stock seuil',
                      label = 'valeur mÃ©diane').options(width = 900, color = 'cyan') #valeur mÃ©diane
        
           
            
        # Renvoie Surplus,PÃ©nurie,Phs,Battery,Methanation,Lake
        
        #DÃ©commenter pour mÃ©thode 1 (intervalles de confiance)
        s,p=StratStockagev3(prodresiduelle,8760,P,B,M,L,T,N,
                                        certitude_interval0,certitude_interval2,certitude_interval1)
        
        #DÃ©commenter pour mÃ©thode 2 (recherche itÃ©rative du meilleur seuil)
        #s,p,P,B,M,L,T,N=StratStockagev3(prodresiduelle,N,Phs,Battery,Methanation,Lake,Thermal,Nuclear,
        #                               seuils_bot, seuils_mid, seuils_top)
        

        heurespartial=heures[d:f]
        prod_stock=np.array(M.prod+B.prod+P.prod+L.prod+T.prod+N.prod)
        prod_Meth=np.array(M.prod)
        prod_Phs=np.array(P.prod)       
        prod_Lake=np.array(L.prod)
        prod_Battery=np.array(B.prod)
        prod_Thermal=np.array(T.prod)
        prod_Nuclear=np.array(N.prod)
        
        
        
        x1 = np.ones(365)*(-1)
        y1 = np.ones(365)*(-1)
        x2 = np.ones(365)*(-1)
        y2 = np.ones(365)*(-1)
        x3 = np.ones(365)*(-1)
        y3 = np.ones(365)*(-1)
        StockPB = P.stored + B.stored ##valeur des deux stocks
        for h in range(24):
            for jour in range(365): ##on regarde tous les jours de l'annÃ©e
                stockage_PB[jour]=StockPB[jour*24+h] #Au jour jour, valeur du stock Phs + Battery
                ##on regarde dans les 24h qui suivent si il y a Ã©crÃªtage, pÃ©nurie ou aucun des deux
                for heure in range(24): 
                    t = (jour * 24 + h + heure) % 8760
                    if s[t] > 0 and StockPB[t] == stockmax : ##cas Ã©crÃªtage
                        x1[jour]=jour+1 ##on note le jour prÃ©cÃ¨dant jour avec Ã©crÃªtage
                        y1[jour]=stockage_PB[jour] ##on note le stock du jour prÃ©cÃ¨dant jour avec Ã©crÃªtage
                    elif p[t] > 0 : ##cas pÃ©nurie
                        x2[jour]=jour+1 ##mÃªmes explications mais pour pÃ©nurie
                        y2[jour]=stockage_PB[jour]
                    else : ##cas ni Ã©crÃªtage, ni pÃ©nurie
                        x3[jour]=jour+1 ##mÃªmes explications mais avec ni Ã©crÃªtage, ni pÃ©nurie
                        y3[jour]=stockage_PB[jour]
                    if x1[jour]==x2[jour]: ##si Ã©cretage et pÃ©nurie le mÃªme jour, on considÃ¨re que c'est une pÃ©nurie 
                        x1[jour]=-1
                        y1[jour]=-1
            ##Nuages de points indiquant si il y a Ã©crÃªtage, pÃ©nurie ou aucun des deux
            I1 = hv.Points((x1[x1!=-1],y1[y1!=-1]), kdims=['j', 'valeur stock Phs + Battery (en GW)'],
                           label = 'Ã©cretage').options(width = 900, color = 'orange', size = 14)
            I2 = hv.Points((x2[x2!=-1],y2[y2!=-1]), kdims=['j', 'valeur stock Phs + Battery (en GW)'],
                           label = 'pÃ©nurie').options(width = 900, color = 'blue', size = 10)
            I3 = hv.Points((x3[x3!=-1],y3[y3!=-1]), kdims=['j', 'valeur stock Phs + Battery (en GW)'],
                           label = 'ok').options(width = 900, color = 'green', size = 6, marker = 'square')
            ##Intervalles de confiance --> distributions pÃ©nurie et Ã©crÃªtage supposÃ©es normale ?
            certitude_interval = certitudeglobal(y1,y2,y3)
            ##Affichage de la zone de confiance sur le graphe : Nuages de points                               
            I4 = hv.Curve((x,certitude_interval[0]*np.ones(365)),
                          label = 'seuil de certitude inf').options(width = 900, color='red')#borne inf
            I5 = hv.Curve((x, certitude_interval[1]*np.ones(365)),
                          label = 'seuil de certitude sup').options(width=900, color = 'black') #borne sup
            I6 = hv.Curve((x, certitude_interval[2]*np.ones(365)),
                          label = 'valeur mÃ©diane').options(width = 900, color = 'cyan') #valeur mÃ©diane
        
        

        
        H8 = hv.Curve((x0, M.stored), "heures", "stock de gaz", label="stock gaz opti").options(width=900)
        
        
        C5=hv.Area((heurespartial,-prod_stock[d:f])).options(width=600)
        
        A1=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore),label='offshore')
        
        A2=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                    +np.array(prod2006_onshore[d:f])*self.onshore),label='onshore')
        
        A3=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv),label='photovoltaic')
        
        A4=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]),label='river')
        
        A5=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]+prod_Battery[d:f]),label='Battery')

        A6=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]+prod_Battery[d:f]\
                   +prod_Phs[d:f]),label='Decharge Phs')

        A7=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]+prod_Battery[d:f]\
                   +prod_Phs[d:f]+prod_Lake[d:f]),label='Lakes')
        
        
        A8=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]+prod_Battery[d:f]\
                   +prod_Phs[d:f]+prod_Lake[d:f]+prod_Meth[d:f]),label='Decharge Methanation')
        
        A9=hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]+prod_Battery[d:f]\
                   +prod_Phs[d:f]+prod_Lake[d:f]+prod_Meth[d:f]+prod_Thermal[d:f]),label='Complement Thermique')
        
        A10 = hv.Area((heurespartial,np.array(prod2006_offshore[d:f])*self.offshore\
                   +np.array(prod2006_onshore[d:f])*self.onshore\
                   +np.array(prod2006_pv[d:f])*self.pv+coefriv*rivprod[d:f]+prod_Battery[d:f]\
                   +prod_Phs[d:f]+prod_Lake[d:f]+prod_Meth[d:f]+prod_Thermal[d:f]\
                   +prod_Nuclear[d:f]),label='Complement Nucleaire')
        
        C4=hv.Curve((heures[d:f],Conso[d:f])).options(width=900,color='black')
        
        #----------------------------------------------------------------------
        
        nbS = 0
        nbP = 0
        
        for i in range(len(s)):
            if s[i] > 0:
                nbS += 1
            if p[i] > 0:
                nbP += 1
                        
        prod_onshore = np.array(prod2006_onshore)
        prod_offshore = np.array(prod2006_offshore)
        prod_pv = np.array(prod2006_pv)

            
        #----------------------------------------------------------------------
                                         
        return pn.Column(C1,
                         C2*C3,
                         G1*G2*G3,
                         H1*H2*H3*H4*H5*H6,
                         I1*I2*I3*I4*I5*I6,
                         H7*H8,
                         (A10*A9*A8*A7*A6*A5*A4*A3*A2*A1*C4).opts(shared_axes=False),
                         (hv.Curve((heures[d:f],prodresiduelle[d:f]),
                                  label='Residual Production').options(width=900,color='black')*C5).opts(shared_axes=False),
                         
                         hv.Curve((heurespartial,P.stored[d:f]+B.stored[d:f]),
                                  label='Stock Phs+Battery').options(width=900, shared_axes=False)) 
```
